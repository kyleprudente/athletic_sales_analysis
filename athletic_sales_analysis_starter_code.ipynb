```python
# Import Libraries and Dependencies
import pandas as pd
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
sales_2020 = pd.read_csv('Resources/athletic_sales_2020.csv')
sales_2021 = pd.read_csv('Resources/athletic_sales_2021.csv')
```

```python
# Display the 2021 sales DataFrame
sales_2021.head()
```

```python
# Check the 2020 sales data types.
sales_2020.dtypes
```

```python
# Check the 2021 sales data types.
sales_2021.dtypes
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_sales = pd.concat([sales_2020,sales_2021],axis='rows',join='inner')
combined_sales.reset_index(drop=True)
```

```python
# Check if any values are null.
combined_sales.count()
```

```python
# Check the data type of each column
combined_sales.dtypes
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_sales = combined_sales.copy()
combined_sales['invoice_date']=pd.to_datetime(combined_sales['invoice_date'])
combined_sales
```

```python
# Confirm that the "invoice_date" data type has been changed.
combined_sales.dtypes
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".
most_products_sold_group = combined_sales.groupby(['region','state','city'])[['units_sold']].sum()
most_products_sold_group = most_products_sold_group.rename(columns={'units_sold':'Total_Products_Sold'})


# Show the top 5 results.
most_products_sold_group = most_products_sold_group.sort_values(by='Total_Products_Sold',ascending=False)
most_products_sold_group.head()
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
most_products_sold_pivot = pd.pivot_table(
    combined_sales,
    index=['region','state','city'],
    values='units_sold',
    aggfunc='sum'
)

# Rename the "units_sold" column to "Total_Products_Sold"
most_products_sold_pivot = most_products_sold_pivot.rename(columns={'units_sold':'Total_Products_Sold'})

# Show the top 5 results.
most_products_sold_pivot = most_products_sold_pivot.sort_values(by='Total_Products_Sold',ascending=False)
most_products_sold_pivot.head()
```

### 5. Determine which Retailer Sold the Most Women's Athletic Footwear

```python
# Filter the sales data to get the women's athletic footwear sales data.
womens_df = combined_sales.loc[combined_sales['product']=="Women's Athletic Footwear"]
womens_df
```

#### Using `groupby`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_retailer_group = womens_df.groupby(['retailer','region','state','city'])[['units_sold']].sum()
women_retailer_group = women_retailer_group.rename(columns={'units_sold':"Womens_Footwear_Units_Sold"})
# Show the top 5 results.
women_retailer_group = women_retailer_group.sort_values(by="Womens_Footwear_Units_Sold",ascending=False)
women_retailer_group.head()
```

#### Using `pivot_table`

```python
# Show the total number of women's athletic footwear sold for each retailer, region, state, and city.
women_retailer_pivot=pd.pivot_table(
    data=womens_df,
    index=['retailer','region','state','city'],
    values='units_sold',
    aggfunc='sum'
)

# Rename the "units_sold" column to "Womens_Footwear_Units_Sold"
women_retailer_pivot = women_retailer_pivot.rename(columns={'units_sold':"Womens_Footwear_Units_Sold"})
# Show the top 5 results.
women_retailer_pivot = women_retailer_pivot.sort_values(by="Womens_Footwear_Units_Sold",ascending=False)
women_retailer_pivot.head()
```

### 5. Determine the Day with the Most Women's Athletic Footwear Sales

```python
# Create a pivot table with the 'invoice_date' column is the index, and the "total_sales" as the values.
day_pivot=pd.pivot_table(
    data=womens_df,
    index='invoice_date',
    values='total_sales',
    aggfunc='sum'
)

# Optional: Rename the "total_sales" column to "Total Sales"
day_pivot = day_pivot.rename(columns={'total_sales':'Total Sales'})

# Show the table.
day_pivot
```

```python
# Resample the pivot table into daily bins, and get the total sales for each day.
daily= day_pivot.resample('D').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
daily = daily.sort_values(by='Total Sales',ascending=False)
daily.head(10)
```

### 6.  Determine the Week with the Most Women's Athletic Footwear Sales

```python
# Resample the pivot table into weekly bins, and get the total sales for each week.
weekly=day_pivot.resample('W').sum()

# Sort the resampled pivot table in ascending order on "Total Sales".
weekly = weekly.sort_values('Total Sales',ascending=False)
weekly.head(10)
```
